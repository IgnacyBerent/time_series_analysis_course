```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf
from statsmodels.graphics.tsaplots import plot_acf
from pandas.plotting import lag_plot
```

```python
df1 = pd.read_csv('AirPassengers.csv', index_col='Month', parse_dates=True)
df1.index.freq = 'MS'
df1.plot()
```

```python
lag_plot(df1['#Passengers'])
```

```python
fig = plot_acf(df1, lags=40, zero=False)
```

```python
diff_df1 = df1.diff().dropna()
diff_df1.plot()
```

```python
fig=plot_acf(diff_df1, lags=40, zero=False)
```

```python
nonsesonal_df1 = diff_df1.diff(12).dropna()
nonsesonal_df1.plot()
```

```python
fig = plot_acf(nonsesonal_df1, lags=40, zero=False)
```

```python
from statsmodels.stats.diagnostic import acorr_ljungbox
```

```python
acorr_ljungbox(nonsesonal_df1, lags=20)
```
