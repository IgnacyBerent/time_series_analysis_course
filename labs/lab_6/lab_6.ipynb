```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import sys
sys.path.append('..')
from lab_5.ma_model import ma_series
from lab_5.ar_model import ar_series
```

# AR(2) Model fit

```python
ar_2, ar_2_formula = ar_series(500, 5000, 8, np.array([1.3, -0.7]))
print(ar_2_formula)
```

```python
plt.plot(ar_2)
```

```python
arima_1 = ARIMA(ar_2, order=(1, 0, 0)).fit()
print(arima_1.summary())
plt.plot(ar_2, label='AR(2)')
plt.plot(arima_1.fittedvalues, label='ARIMA(1, 0, 0)')
plt.legend()
plt.show()
```

```python
arima_2 = ARIMA(ar_2, order=(2, 0, 0)).fit()
print(arima_2.summary())
plt.plot(ar_2, label='AR(2)')
plt.plot(arima_2.fittedvalues, label='ARIMA(2, 0, 0)')
plt.legend()
plt.show()
```

```python
arima_3 = ARIMA(ar_2, order=(3, 0, 0)).fit()
print(arima_3.summary())
plt.plot(ar_2, label='AR(2)')
plt.plot(arima_3.fittedvalues, label='ARIMA(3, 0, 0)')
plt.legend()
plt.show()
```

```python
arima_4 = ARIMA(ar_2, order=(4, 0, 0)).fit()
print(arima_4.summary())
plt.plot(ar_2, label='AR(2)')
plt.plot(arima_4.fittedvalues, label='ARIMA(4, 0, 0)')
plt.legend()
plt.show()
```

Difference between preddictions of ARIMA(2,0,0) and ARMIA(3,0,0) and ARIMA(4,0,0) models isn't significant, and AIC of ARIMA(2,0,0) is the smallest.

# MA(2) Model fit

```python
ma_2, ma_2_formula = ma_series(500, 5000, 0, np.array([-1, 0.8]))
print(ar_2_formula)
```

```python
plt.plot(ma_2)
```

```python
moving_average_1 = ARIMA(ma_2, order=(0, 0, 1)).fit()
print(moving_average_1.summary())
plt.plot(ma_2, label='MA(2)')
plt.plot(moving_average_1.fittedvalues, label='ARIMA(0, 0, 1)')
plt.legend()
plt.show()
```

```python
moving_average_2 = ARIMA(ma_2, order=(0, 0, 2)).fit()
print(moving_average_2.summary())
plt.plot(ma_2, label='MA(2)')
plt.plot(moving_average_2.fittedvalues, label='ARIMA(0, 0, 2)')
plt.legend()
plt.show()
```

```python
moving_average_3 = ARIMA(ma_2, order=(0, 0, 3)).fit()
print(moving_average_3.summary())
plt.plot(ma_2, label='MA(2)')
plt.plot(moving_average_3.fittedvalues, label='ARIMA(0, 0, 3)')
plt.legend()
plt.show()
```

```python
moving_average_4 = ARIMA(ma_2, order=(0, 0, 4)).fit()
print(moving_average_4.summary())
plt.plot(ma_2, label='MA(2)')
plt.plot(moving_average_4.fittedvalues, label='ARIMA(0, 0, 4)')
plt.legend()
plt.show()
```

The same rule applies to the MA models. AIC of ARIMA(0,0,2) is the smallest.
