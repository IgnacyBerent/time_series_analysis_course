```python
import numpy as np
import matplotlib.pyplot as plt

from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller

from ma_model import ma_series
from ar_model import ar_series
```

# Constants setup

```python
BURNIN = 500
N = 5000
SEED = 42
```

# Correlations in AR model

```python
ar_series_1, ar_formula_1 = ar_series(burnin=BURNIN, n=N, c=18, o=np.array([-0.6]), seed=SEED)

plt.title(ar_formula_1)
plt.plot(ar_series_1)
fig = plot_acf(ar_series_1, lags=20, zero=False)
fig = plot_pacf(ar_series_1, lags=20, zero=False)
acorr_ljungbox(ar_series_1, lags=20, return_df=True)
```

```python
ar_series_2, ar_formula_2 = ar_series(burnin=BURNIN, n=N, c=18, o=np.array([-0.7]), seed=SEED)

plt.title(ar_formula_2)
plt.plot(ar_series_2)
fig = plot_acf(ar_series_2, lags=20, zero=False)
fig = plot_pacf(ar_series_2, lags=20, zero=False)
acorr_ljungbox(ar_series_2, lags=20, return_df=True)
```

## Is stationary if |theta| > 1?

```python
ar_series_5, ar_formula_5 = ar_series(burnin=BURNIN, n=N, c=18, o=np.array([0.999]), seed=SEED)

plt.title(ar_formula_5)
plt.plot(ar_series_5)
fig = plot_acf(ar_series_5, lags=20, zero=False)
fig = plot_pacf(ar_series_5, lags=20, zero=False)
```

```python
ar_series_5, ar_formula_5 = ar_series(burnin=BURNIN, n=N, c=18, o=np.array([1.0]), seed=SEED)

plt.title(ar_formula_5)
plt.plot(ar_series_5)
fig = plot_acf(ar_series_5, lags=20, zero=False)
fig = plot_pacf(ar_series_5, lags=20, zero=False)
```

We can observe that if theta get closer to one then the series becomes more and more non stationary and even becomes a linear function.

# Comparison of AR and MA models

We can observe that ar models tend to have bigget autocorrelation and low partial autocorrelation. On the other hand, ma models behave the opposite way.

From the above we can conclude that usage of AR models is more suitable if we want to capture the direct influence of previous values on the current one. On the other hand, MA models are better in capturing the effects of past points without considerating possible fluctuations from closest points.
